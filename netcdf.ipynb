```python
import netCDF4 as nc

# Set the path to the NetCDF file
nc_file = '/content/drive/MyDrive/ECMWF_ERA-40_subset.nc'

# Open the NetCDF file for reading
with nc.Dataset(nc_file, 'r') as ds:
    
    # Get the variable attributes
    variable_name = 'hcc'  # replace 't2m' with the variable name you want to get the long name for
    var_attrs = ds.variables[variable_name].__dict__

    # Print the long name
    if 'long_name' in var_attrs:
        print(f"The long name of '{variable_name}' is '{var_attrs['long_name']}'")
    else:
        print(f"No long name found for '{variable_name}'")
```

```python
import netCDF4 as nc

# Set the path to the NetCDF file
nc_file = '/content/drive/MyDrive/ECMWF_ERA-40_subset.nc'

    
   # List of variables to retrieve long_name attribute for
var_list = ['longitude', 'latitude', 'time', 'tcw', 'tcwv', 'lsp', 'cp', 'msl', 'blh', 'tcc', 'p10u', 'p10v', 'p2t', 'p2d', 'e', 'lcc', 'mcc', 'hcc', 'tco3', 'tp']

# Open the NetCDF file for reading
with nc.Dataset(nc_file, 'r') as ds:
    
    # Loop over variables and print their long_name attribute
    for var_name in var_list:
        var = ds.variables[var_name]
        print(f"{var_name}: {var.long_name}")
```

```python
import netCDF4 as nc

# Mount Google Drive to access the file
from google.colab import drive
drive.mount('/content/drive')

# Set the path to the NetCDF file
nc_file = '/content/drive/MyDrive/ECMWF_ERA-40_subset.nc'

# Open the NetCDF file for reading
with nc.Dataset(nc_file, 'r') as ds:
    
    # Get the list of variable names
    var_list = ds.variables.keys()

    # Loop over each variable name
    for var_name in var_list:
        
        # Get the long name for the variable
        long_name = ds.variables[var_name].long_name
        
        # Print the variable name and long name
        print(var_name + ': ' + long_name)
```
